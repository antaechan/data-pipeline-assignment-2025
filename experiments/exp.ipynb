```python
import pandas as pd
import os

df = pd.read_csv("../data/cell2info.csv")
```

```python
primary_key = "CellID"

not_primary_columns = [c for c in df.columns if c != primary_key]
not_primary_columns
```

```python
for i, c1 in enumerate(not_primary_columns):
    for j, c2 in enumerate(not_primary_columns):
        if i != j:
            key = df[c1].dropna().unique()
            temp_df = df[[c1,c2]]
            print(f"# keys in {c1}: {key.shape[0]}")
            temp_df = temp_df[temp_df[c1].isin(key)]
            temp_df = temp_df.drop_duplicates()
            print(temp_df.shape[0])
            print(f"# Nan in {c1}: {temp_df[c1].isna().sum()}")
            print(f"# Nan in {c2}: {temp_df[c2].isna().sum()}")
            print(temp_df)
            print()
            
```
