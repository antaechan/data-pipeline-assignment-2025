```python
import pandas as pd
from itertools import combinations

# 기본 함수 종속성을 추정하는 함수
def find_partial_dependencies(df):
    columns = df.columns.to_list()
    partial_dependencies = []
    
    # 기본 키 후보 추정: 모든 가능한 조합을 기본 키 후보로 봄
    # 이 예시는 간단히 모든 2개 이상의 컬럼 조합을 기본 키 후보로 간주합니다
    # (현실적인 경우 기본 키 후보를 추정하는 방법은 더 복잡합니다)
    for n in range(1, len(columns) + 1):
        for pk_candidate in combinations(columns, n):
            # 각 속성에 대해 부분 함수 종속성을 찾습니다
            # pk_candidate 기본 키 후보
            for col in columns:
                if col not in pk_candidate:
                    # combo에 포함되지 않은 컬럼이 해당 기본 키의 일부에 종속되었는지 확인
                    # 부분 함수 종속: combo의 일부에만 종속되는지 확인
                    grouped = df.groupby(list(pk_candidate))[col].nunique()
                    # print("pk_candidate: ", pk_candidate)
                    # print("col: ", col)
                    # print(grouped)
                    

                    # (3) 대상 컬럼(col)이 기본 키 후보(copk_candidatembo)의 일부에만 종속되는지 체크
                    if grouped.max() <= 1:
                        partial_dependencies.append((col, pk_candidate))
                        

    return partial_dependencies


# 부분 함수 종속성 찾기
partial_dependencies = find_partial_dependencies(df)

# 결과 출력
for col, pk_candidate in partial_dependencies:
    print(f"부분 함수 종속성: {pk_candidate} <- {col}")
```
